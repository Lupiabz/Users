```python
import os

from modules.setting import (
    chara_folders,
    aug_folders,
    CHARACTER_PATH,
    BASE_PATH,
    OUTPUT_PATH,
)
```

```python
def ask_make_dir():
    """
    フォルダを作成するかどうかy/nで入力
    """
    while True:
        # フォルダを作成するかどうかy/nで入力
        make_dir = input(f"{BASE_PATH} にフォルダを作成しますか？(y/n): ").lower()
        if make_dir == "y":
            if not os.path.exists(BASE_PATH):
                raise FileNotFoundError(f"{BASE_PATH} が存在しません. BASE_PATHを確認してください")
            return True
        elif make_dir == "n":
            return False
        else:
            print("yかnで入力してください")
```

```python
def make_dirs():
    if ask_make_dir():
        print("フォルダを作成します")
        print(f"作成先: {BASE_PATH}/")
        os.makedirs(f"{OUTPUT_PATH}", exist_ok=True)
        for folder in chara_folders:
            os.makedirs(f"{CHARACTER_PATH}/{folder}", exist_ok=True)
        for folder in aug_folders:
            os.makedirs(f"{CHARACTER_PATH}/{folder}", exist_ok=True)
        os.makedirs(f"{CHARACTER_PATH}/test", exist_ok=True)
        print("フォルダを作成しました")
    else:
        print("フォルダを作成しませんでした")
```

```python
make_dirs()
```
